## metrics

```python
def segm_acc(input, target):
    target = target.squeeze(1)
    return (input.argmax(dim=1)==target).float().mean()

def iou(input:Tensor, targs:Tensor)->Rank0Tensor:
    n = targs.shape[0]
    input = input.argmax(dim=1).view(n,-1)
    targs = targs.view(n,-1)
    intersect = (input == targs).sum().float()
    union = ((input+targs) > 0).sum().float()
    return intersect / (union + 1.0)

def iou_for_class(input:Tensor, targs:Tensor, class_idx)->Rank0Tensor:
    n = targs.shape[0]
    input = input.argmax(dim=1).view(n,-1)
    targs = targs.view(n,-1)
    
    mask_true = (input == class_idx).float()
    mask_pred = (targs == class_idx).float()
    iou = (mask_pred * mask_true).sum() / (mask_true.numel() - ((mask_pred + mask_true)==0).sum() + 1)
    
    return iou.mean()

def precision_for_class(input:Tensor, targs:Tensor, class_idx)->Rank0Tensor:
    n = targs.shape[0]
    input = input.argmax(dim=1).view(n,-1)
    targs = targs.view(n,-1)
    
    mask_true = (input == class_idx).float()
    mask_pred = (targs == class_idx).float()
    precision = ((mask_pred * mask_true).sum() / 
           ((mask_pred * (1 - mask_true)).sum() + (mask_pred * mask_true).sum() + 1))
    
    return precision.mean()

def recall_for_class(input:Tensor, targs:Tensor, class_idx)->Rank0Tensor:
    n = targs.shape[0]
    input = input.argmax(dim=1).view(n,-1)
    targs = targs.view(n,-1)
    
    mask_true = (input == class_idx).float()
    mask_pred = (targs == class_idx).float()
    recall = ((mask_pred * mask_true).sum() / 
           (((1 - mask_pred) * mask_true).sum() + (mask_pred * mask_true).sum() + 1))
    
    return recall.mean()

def f1_score_for_class(input:Tensor, targs:Tensor, class_idx)->Rank0Tensor:
    precision = precision_for_class(input, targs, class_idx)
    recall = recall_for_class(input, targs, class_idx)
    
    f1 = 2 * precision * recall / (precision + recall)
    return f1
```

```python
metrics = [
    [partial(metric, class_idx=class_idx) 
      for class_idx in [0, 1, 2]
    ] 
    for metric in [iou_for_class, precision_for_class, recall_for_class, f1_score_for_class]
] 
    
metrics = [segm_acc, iou] + [item for sublist in metrics for item in sublist]
```

```python
wd=1e-2
learn = unet_learner(data,
                     models.resnet34,
                     metrics=metrics,
                     wd=wd
                    )

path_to_model = Path('/home/data/fastai_data/data_split_merged_new/models/big-5-epochs')
learn = learn.load(path_to_model)
```
