## metrics

```python
def segm_acc(input, target):
    target = target.squeeze(1)
    return (input.argmax(dim=1)==target).float().mean()

def iou(input:Tensor, targs:Tensor)->Rank0Tensor:
    n = targs.shape[0]
    input = input.argmax(dim=1).view(n,-1)
    targs = targs.view(n,-1)
    intersect = (input == targs).sum().float()
    union = ((input+targs) > 0).sum().float()
    return intersect / (union + 1.0)

def iou_for_class(input:Tensor, targs:Tensor, class_idx)->Rank0Tensor:
    n = targs.shape[0]
    input = input.argmax(dim=1).view(n,-1)
    targs = targs.view(n,-1)
    
    mask_true = (input == class_idx).float()
    mask_pred = (targs == class_idx).float()
    iou = (mask_pred * mask_true).sum() / (mask_true.numel() - ((mask_pred + mask_true)==0).sum() + 1)
    
    return iou.mean()
```

```python
metrics = [
   segm_acc,
   iou,
   partial(iou_for_class, class_idx=0),
   partial(iou_for_class, class_idx=1),
   partial(iou_for_class, class_idx=2),
]
```
